```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#upload asin list
asins = pd.read_excel('./data/Top300.xlsx')
links = asins['Asin'].to_list()
```

```python
prices = []

for link in links: 
    url = f'https://www.amazon.com/dp/{link}'
    page = requests.get(url)

    soup = BeautifulSoup(page.content, 'html.parser')
    try: 
        price = soup.find_all('span', {'class': 'a-offscreen'})[0].get_text().replace('Page 1 of 1', 'Currently Unavailable')
    except: 
        price = 'Page Not Found ʢ・●・ʡ'

    prices.append(price)


print(prices)
```

```python
asin_price = pd.DataFrame({
    'asin' : links, 
    'price' : prices}, 
    columns=['asin', 'price'])

asin_price.to_excel('asin_price_list.xlsx')
```
